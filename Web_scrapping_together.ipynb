```python
platform = 'NS'
last_page = 24
for p in range(1,last_page):
    url = f"https://www.vgchartz.com/games/games.php?page={p}&console={platform}&order=Sales&ownership=Both&direction=DESC&showtotalsales=1&shownasales=0&showpalsales=0&showjapansales=0&showothersales=0&showpublisher=1&showdeveloper=0&showreleasedate=1&showlastupdate=0&showvgchartzscore=1&showcriticscore=1&showuserscore=1&showshipped=1"
    html_doc = pd.read_html(url)    
    #Grab the columns on the page 
    df = html_doc[-2]
    if p == 1:
        dfA = pd.DataFrame()
        dfB = pd.DataFrame()
        dfA['Names'] = df.iloc[:,2]
        dfA['Total_shipped'] = df.iloc[:,8]
        dfA['Total_sales'] = df.iloc[:,9]
    elif p == last_page :
        dfB = pd.DataFrame()
        dfB['Names'] = df.iloc[:,2]
        dfB['Total_shipped'] = df.iloc[:,8]
        dfB['Total_sales'] = df.iloc[:,9]
        dfA = pd.concat([dfA, dfB])
        
    elif p > 1 :
        dfB['Names'] = df.iloc[:,2]
        dfB['Total_shipped'] = df.iloc[:,8]
        dfB['Total_sales'] = df.iloc[:,9]
        dfA = pd.concat([dfA, dfB])
```

```python
#With 3DS
if platform == "3DS":
    df_3DS = dfA
    df_3DS.reset_index(inplace = True)
    df_3DS.to_csv() 
#With iOS
elif platform == "iOS":
    df_iOS = dfA
    df_iOS.reset_index(inplace = True)
elif platform == "NS":
    df_NS = dfA
    df_NS.reset_index(inplace = True)
```
